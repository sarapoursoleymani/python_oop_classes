    <h4>Question 01 To Do List OOP Project(300 + 50):</h4>
    <p>You are tasked with creating a ToDo list manager using Object-Oriented Programming (OOP).
        Your program should consist of atleastthree classes: Task, Manager, and a Main function for
        running the application. Each class should be in a separate file to manage the project more
        effectively
    </p>
<p>Task Class:</p>
<ul>
<li>Each task must have the following attributes:</li>
<ul>
<li>A title (string).</li>
<li>A description(string).</li>
<li>A unique task ID, which should be a number but must be stored as a string (your program
should check thatthe ID is numeric).</li>
<li>Starttime, which is automatically recorded when a task is created.</li>
<li>End time, which is recorded when the task is marked as done.</li>
<li>Status: A boolean value thatis setto False when the task is created and setto True when the
task is marked as done.</li>
</ul>
</ul>
<p>Manager Class</p>
<ul>
<li>The manager should be able to:</li>
<ul>
<li>Add a task: Create and add a new task with a unique task ID.</li>
<li>Change task status to done: Mark the task as completed, update the status to True, and
record the end time.
</li>
<li>Display tasks: Show alltasks along with their details, such as task ID,title, description,
starttime, end time (if completed), and current status (done or not done).</li>
<li>Task summary: Display the number oftasks that are completed (status = True) and those
that are not completed (status = False).
</li>
</ul>
</ul>
<p>Main:</p>
<ul>
<li>The manager should be able to:</li>
<ul>
<li>The main function should interact with the Managerto allow users to perform actions
such as adding a task, marking a task as done, and viewing alltasks.</li>
</ul>
</ul>
<p style = "color:black;">Additional Requirements:</p>
<ul>
<li>Your program must be splitinto atleastthree files: one forthe Task class, one forthe Manager class, and
one forthe Main.</li>
<li>Make sure task IDs are unique</li>
<li>Ensure your program can handle and display the task details correctly.</li>
</ul>
</div>

```python
import time
class Task:
    def __init__(self, task_id: str, title: str, description: str):
        if not task_id.isdigit():
            raise ValueError("Task ID باید فقط عددی باشد")
        self.task_id = task_id
        self.title = title
        self.description = description
        self.start_time = time.strftime("%Y-%m-%d %H:%M:%S", time.localtime())
        self.end_time = None
        self.status = False 
    def mark_done(self):
        self.status = True
        self.end_time = time.strftime("%Y-%m-%d %H:%M:%S", time.localtime())
    def __str__(self):
        return (f"Task ID: {self.task_id}, Title: {self.title}, Description: {self.description},Status: {'Done' if self.status else 'Not Done'}, Start: {self.start_time},End: {self.end_time if self.end_time else 'N/A'}")
class Manager:

    def __init__(self):
        self.tasks = {}

    def add_task(self, task_id, title, description):
        if task_id in self.tasks:
            raise ValueError(f"Task ID {task_id} از قبل وجود دارد")
        new_task = Task(task_id, title, description)
        self.tasks[task_id] = new_task
    def mark_task_done(self, task_id):
        if task_id not in self.tasks:
            raise ValueError(f"Task با ID {task_id} یافت نشد")
        self.tasks[task_id].mark_done()
    def display_tasks(self):
        for task_id, task in self.tasks.items():
            print(task)
    def task_summary(self):
        done_tasks = sum(1 for task in self.tasks.values() if task.status)
        not_done_tasks = len(self.tasks) - done_tasks
        print(f"Tasks Done: {done_tasks}, Tasks Not Done: {not_done_tasks}")
    def main():
      manager = Manager()
      while True:
        print("1. افزودن وظیفه جدید")
        print("2. علامت زدن وظیفه به عنوان انجام شده")
        print("3. نمایش تمامی وظایف")
        print("4. نمایش خلاصه وظایف")
        print("5. خروج")
        choice = input("انتخاب کنید: ")
        if choice == '1':
            task_id = input("شناسه وظیفه: ")
            title = input("عنوان: ")
            description = input("توضیحات: ")
            try:
                manager.add_task(task_id, title, description)
                print("وظیفه با موفقیت اضافه شد.")
            except ValueError as e:
                print(e)
        elif choice == '2':
            task_id = input("شناسه وظیفه برای علامت‌گذاری: ")
            try:
                manager.mark_task_done(task_id)
                print("وظیفه به عنوان انجام شده علامت‌گذاری شد.")
            except ValueError as e:
                print(e)
        elif choice == '3':
            print("لیست وظایف:")
            manager.display_tasks()
        elif choice == '4':
            print("خلاصه وظایف:")
            manager.task_summary()
        elif choice == '5':
            print("خروج از برنامه.")
            break
        else:
            print("انتخاب نامعتبر. لطفاً دوباره انتخاب کنید.")

    if __name__ == "__main__":
        main()
```

    <h4>Question 01(Inheritance)(200 + 50)</h4>
<p>You are tasked with creating a system for managing electronic devices. Both Laptops and
Smartphones share some common features, butthey also have their own specific
characteristics. Use inheritance to avoid repeating code.<p>
<p>Device Class (Base Class):</p>
<ul>
Common Attributes (shared by both laptops and smartphones):
<ul>
<li>Brand (string)</li>
<li>Model(string)</li>
<li>Price(float)</li>
</ul>
</ul>
<ul>
Common Methods:
<ul>
<li>Turn On: Prints a message thatthe device is powered on.</li>
<li>Turn Off: Prints a message thatthe device is powered off.</li>
<li>Display Info: Displays the brand, model, and price ofthe device.
</li>
</ul>
</ul>
<ul>
Laptop Class (Derived Class):
<ul>
<li>Inherits all attributes and methods from the Device class.
</li>
<li>Additional Attribute:</li>
<ul><li>RAM Size (integer,GB)</li></ul>
<li>Additional Method:</li>
<ul><li>Open Laptop: Prints a message indicating the laptop is opened.</li></ul>
</ul>
</ul>
<ul>
Smartphone Class (Derived Class):
<ul>
<li>Inherits all attributes and methods from the Device class.
</li>
<li>Additional Attribute:</li>
<ul><li>Camera Resolution (integer, MP)</li></ul>
<li>Additional Method:</li>
<ul><li>Take Photo: Prints a message saying a photo has been taken.</li></ul>
</ul>
</ul>
</div>

```python
class Device:
    def __init__(self, brand, model, price):
        self.brand = brand
        self.model = model
        self.price = price
    def turn_on(self):
        print(f"{self.brand} {self.model} is now powered on.")
    def turn_off(self):
        print(f"{self.brand} {self.model} is now powered off.")
    def display_info(self):
        print(f"Brand: {self.brand}, Model: {self.model}, Price: {self.price}")
class Laptop(Device):
    def __init__(self, brand, model, price, ram_size):
        super().__init__(brand, model, price)
        self.ram_size = ram_size
    def open_laptop(self):
        print(f"{self.brand} {self.model} laptop is now opened.")
    def display_info(self):
        super().display_info()
        print(f"RAM Size: {self.ram_size} GB")
class Smartphone(Device):
    def __init__(self, brand, model, price, camera_resolution):
        super().__init__(brand, model, price)
        self.camera_resolution = camera_resolution
    def take_photo(self):
        print(f"{self.brand} {self.model} has taken a photo with {self.camera_resolution} MP camera.")
    def display_info(self):
        super().display_info()
        print(f"Camera Resolution: {self.camera_resolution} MP")
laptop = Laptop("ASUS", "core i7", 1620.00, 14)
smartphone = Smartphone("Apple", "iPhone 13", 975.00, 10)
laptop.turn_on()
laptop.display_info()
laptop.open_laptop()
laptop.turn_off()
print()
smartphone.turn_on()
smartphone.display_info()
smartphone.take_photo()
smartphone.turn_off()
```
