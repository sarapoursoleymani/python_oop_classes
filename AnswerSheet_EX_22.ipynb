    <h4>Question 01 (250):</h4>
    <p>Create a digital store that manages two types of products: Phone and Laptop. Each product type has shared attributes, including an ID, name, price, and brand. Additionally, Laptop has specific attributes for memory and processor. In this project, you will define three classes: Phone, Laptop, and Store.</p> 
<p>Class Phone</p>
<li>Attributes:</li>
<li>id (str): The unique identifier of the product.</li>
<li>name (str): The name of the product.</li>
<li>price (float): The price of the product.</li>
<li>brand (str): The brand of the product.</li> 
<p>Methods:</p>
<li>display(): Display the product information.</li>

<p>Class Laptop</p>
<li>Attributes:</li>
<li>id (str): The unique identifier of the product.</li>
<li>name (str): The name of the product.</li>
<li>price (float): The price of the product.</li>
<li>brand (str): The brand of the product.</li> 
<li>memory (str): The memory of the laptop.</li>
<li>processor (str): The processor type of the laptop</li>
<p>Methods:</p>
<li>display(): Display the product information.</li>

<p>Class Store</p>
<li>Attributes:</li>
<li>phones: A collection to store phones.</li>
<li>laptops: A collection to store laptops.</li> 
<p>Methods:</p>
<li>add(): Add phone or laptop the phones collection.</li>
<li>emove(id): Remove phone or laptop collection using the ID</li>
<li>display(type): Display all phones or all laptops</li>
</div>

```python
class Phone:
    def __init__(self, id, name, price, brand):
        self.id = id
        self.name = name
        self.price = price
        self.brand = brand
    def display(self):
        print(f"Phone ID: {self.id}")
        print(f"Name: {self.name}")
        print(f"Price: ${self.price}")
        print(f"Brand: {self.brand}")
        print("-------------------")
class Laptop:
    def __init__(self, id, name, price, brand, memory, processor):
        self.id = id
        self.name = name
        self.price = price
        self.brand = brand
        self.memory = memory
        self.processor = processor
    def display(self):
        print(f"Laptop ID: {self.id}")
        print(f"Name: {self.name}")
        print(f"Price: ${self.price}")
        print(f"Brand: {self.brand}")
        print(f"Memory: {self.memory}")
        print(f"Processor: {self.processor}")
        print("-------------------")
class Store:
    def __init__(self):
        self.phones = []
        self.laptops = []
    def add(self, product):
        if isinstance(product, Phone):
            self.phones.append(product)
        elif isinstance(product, Laptop):
            self.laptops.append(product)
        else:
            print("Unknown product type")
    def remove(self, id):
        for product in self.phones + self.laptops:
            if product.id == id:
                if isinstance(product, Phone):
                    self.phones.remove(product)
                elif isinstance(product, Laptop):
                    self.laptops.remove(product)
                print(f"Product with ID {id} has been removed.")
                return
        print(f"No product found with ID {id}.")
    def display(self, type):
        if type.lower() == "phone":
            print("Displaying all phones:")
            for phone in self.phones:
                phone.display()
        elif type.lower() == "laptop":
            print("Displaying all laptops:")
            for laptop in self.laptops:
                laptop.display()
        else:
            print("Unknown type. Please choose 'phone' or 'laptop'.")
store = Store()
phone1 = Phone("P001", "iPhone 13", 999.99, "Apple")
phone2 = Phone("P002", "Galaxy S21", 799.99, "Samsung")
laptop1 = Laptop("L001", "MacBook Pro", 1299.99, "Apple", "16GB", "M1")
laptop2 = Laptop("L002", "Dell XPS", 999.99, "Dell", "8GB", "Intel i7")
store.add(phone1)
store.add(phone2)
store.add(laptop1)
store.add(laptop2)
store.display("phone")
store.display("laptop")
store.remove("P002")
store.display("phone")
```
